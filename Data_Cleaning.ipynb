```python
import pandas as pd
```

```python
# Read CSV into Pandas DataFrame
file_path = 'Starting_Data/Project4_StartingData_noMeta.csv'
df = pd.read_csv(file_path)
df
```

```python
# Check Genus Value Counts
value_counts = df['genus'].value_counts()
value_counts_df = pd.DataFrame(value_counts)
value_counts_updated_df = value_counts_df.drop(value_counts_df[value_counts_df.genus < 50].index)
value_counts_updated_df.head(11)
```

```python
# Remove Quotes Function
def remove_quotes(value):
    if isinstance(value, str):
        return value.strip('"')
    return value
```

```python
# Create Clean DataFrame with desired columns
column_df = df[['max_ma', 'min_ma', 'genus', 'lng', 'lat', 'lithology1', 'environment']]
clean_df = column_df.applymap(remove_quotes)
```

```python
# Fill NA values in the lithology1 column with 'not reported' to match other data
clean_df['lithology1'] = clean_df['lithology1'].fillna('not reported')
```

```python
# Check value counts of the lithology1 column
clean_value_counts = clean_df['lithology1'].value_counts()
print(clean_value_counts)
```

```python
# Drop any remaining NA values, run the remove quotes function, and view the DataFrame
clean_df = clean_df.dropna(how='any')
clean_df
```

```python
# Create a new excel file in the directory with the cleaned DataFrame
clean_df.to_excel('Starting_Data/cleaned_data.xlsx', index=False)
```
